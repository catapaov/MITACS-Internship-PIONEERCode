This notebook is used to display the fits of the charge distribution in order to extract the number of photoelectrons from the SPE distribution.

Note that the displayed distributions are not SPEs themselves, but are used just to quickly check how many photoelectrons we are seeing.

The implementation of the functions is in the file SPE_fit.py. The general idea is to change the filename to the desired one and run the function: "compute_area(file, t0, t1)"
Here, t0 is the time at which the PMT pulse starts, and t1 is the time point when the pulse ends. This function computes the area-charge of the pulse.

After that, you must call the function: "plot_and_compute_spe()"

This function performs the fit and displays the result.



```python
import SPE_fit as spe

date= '25-07-24'
PMT_ID = 'BA0100'
v_set= 1100
LED_ID = "308"
t0, t1   = 0.38e-7, 2e-7
```

```python
file= f"/home/aovelencio/PMTTesting/SPE_PMT_data/{date}/CHARGE_CHECK_1e4wf_SPEdataTest_{PMT_ID}-{v_set}V_L-{LED_ID}_LASER-ON_CH1.csv"
```

```python
df=spe.compute_area(file,t0,t1)
areas = df['Area'].values
```

```python
spe.plot_and_compute_spe(areas)
```
